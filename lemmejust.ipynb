```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler 
%matplotlib inline
np.random.seed(0)
plt.rcParams['figure.figsize'] = [20, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
sns.set_theme(context="notebook")
```

```python
test = pd.read_csv('./test.csv', index_col='id')
dT = pd.read_csv('./train.csv', index_col='id')
dT.head()
```

```python
scaler = MinMaxScaler()
scaler.fit(dT)
train = pd.DataFrame(scaler.transform(dT))
train.index = dT.index
train.columns = dT.columns
train.head()
```

```python
scaler2 = MinMaxScaler()
scaler2.fit(test)
test_Scld = pd.DataFrame(scaler2.transform(test))
test_Scld.index = test.index
test_Scld.columns = test.columns
test_Scld.head()
```

```python
from tqdm import tqdm
plt.figure(figsize=(24, 6*(118/4)))
for i in tqdm(range(len(train.columns.tolist())-1)):
    plt.subplot(30, 4, i+1)
    sns.histplot(train[f'f{i+1}'], kde=True)
    sns.histplot(test_Scld[f'f{i+1}'], kde=True, color='red')
plt.show()
```
